```python
import os
import shutil
import random
import pandas as pd
```

```python
# ***** IMPORTANTE!!! *****
# Alterar de acordo com a sua máquina.
# *************************
MAIN_PATH = '/home/bianca/Área de Trabalho/Weeds/'

source_folder = os.path.join(MAIN_PATH, 'dataset')

destination_folder = os.path.join(MAIN_PATH, 'dataset_')


# Cria as subpastas se não existirem
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)


# Lista todas as imagens na pasta de origem
images = [f for f in os.listdir(source_folder) if os.path.isfile(
    os.path.join(source_folder, f))]
random.shuffle(images)

total_images = len(images)
print(total_images)
```

```python
df = pd.read_csv(os.path.join(MAIN_PATH, 'labels.csv'))

print(df.to_string()) 

class_list = df['Species'].tolist()
class_names = list(set(class_list))
num_classes = len(class_names)
print(class_names)
print(num_classes)
```

```python
for class_name in class_names:
    if not os.path.exists(os.path.join(destination_folder, class_name)):
        os.makedirs(os.path.join(destination_folder, class_name))
```

```python
for index, row in df.iterrows():
    print(index, row['Filename'], row['Species'])

    shutil.copy2(os.path.join(source_folder, row['Filename']), os.path.join(destination_folder, row['Species']))
```
